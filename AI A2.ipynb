# K-Means Algorithm



K-Mean is one of the most straightforward and well-known unsupervised learning strategies for clustering issues.

The process uses a predetermined number of clusters that you have selected before you run the algorithm to classify a given data set in a straightforward and straightforward manner. 

You choose the number k, or the number of clusters, at random. Then, for each data point, the k-means algorithm will locate the closest cluster centre and allocate the data point that is closest to that cluster. The cluster centres will be updated once each data point has been assigned to a cluster. 

In other words, as the values of the cluster data points are averaged, new cluster centres will form.Until the majority of the data points remain in the same cluster, this process is repeated. The cluster membership ought to settle down. 

Because of the way the k-means algorithm operates, it is clear that the initial set of cluster centres that are abandoned can have a big impact on the clusters that are ultimately discovered. 

Of course, you can deal with this consequence however, you should absolutely look into getting your data ready for this purpose before you move on to creating a model for it.
